## Converting `price` and `odometer` to numerical values
As we have seen those two fields are stored as text, probaby due to the extra characters in them. We want to make these two columns numerical values. <br>
We are going to start by removing those unnecessary characters, and then converting the column to a numeric type.<br>
We will also rename the columns so they add the extra description needed after the delation of the characters.

### Removing unwanted characters and converting the values in the columns to numeric characters

```python
autos["price"] = autos["price"].str.replace("$","").str.replace(",","").astype(int)
autos["odometer"] = autos["odometer"].str.replace("km","").str.replace(",","").astype(int)
```

## Exploring price by brand

We want to know wich brands are the most reliable in eBay.

```python
autos["brand"].unique()
```

```python
autos["brand"].value_counts(normalize = True)
```

Volkswagen, BMW, Opel, Mercedes_benz, and Audi are the most popular brands in eBay Germany, which makes sense considering they are all German brands.<br>
As such, we are going to select these brands to aggregate them. The fact that they account for 5% of the market is also a good reason to analyses these brands.<br>
So what's the average price of these brands?

```python
top_5_brands_mean = {}

brand = autos["brand"].value_counts(normalize = True)
print(brand)
```

```python
top_5_brands = brand.index[:5]
print(top_5_brands)
```

```python
for b in top_5_brands:
    selected_rows = autos[autos["brand"]== b]
    mean_price_top_5_brands = selected_rows["price_dollars"].mean()
    top_5_brands_mean[b] = mean_price_top_5_brands

top_5_brands_mean  #How would I sort this one by the mean?    
```

Opel is the cheapest of all the brands, followed by volkswagen. While Audi is the most expensive one, even more so than BMW and Mercedes, meaning that maybe Audis keep their value better.

Or could the mileage have something to do with it?

```python
top_5_mean_mileage = {}

for b in top_5_brands:
    selected_rows = autos[autos["brand"]==b]
    mean_mileage = selected_rows["odometer_km"].mean()
    top_5_mean_mileage[b] = mean_mileage

top_5_mean_mileage
```

Let's combine both results into a dataframe so we can see it more clearly

```python
#using a series constructure to convert dictionaries into series
price_mean = pd.Series(top_5_brands_mean)
mileage_mean = pd.Series(top_5_mean_mileage)

#combining the series into a dataframe
price_mileage = pd.DataFrame({"Mean price USD":price_mean,"Mean mileage KM":mileage_mean})
print(price_mileage)
```
